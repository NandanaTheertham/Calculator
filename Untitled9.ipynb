Plot a graph of set size vs no.of subsets

```python
import matplotlib.pyplot as plt
import random

x=[]
y=[]
def find_subsets(s):
    if len(s) == 0:
        return [[]]
    else:
        # Convert the set to a list to allow slicing
        s_list = list(s)
        subsets = find_subsets(s_list[1:])
        return subsets + [[s_list[0]] + subset for subset in subsets]


def create_random_set(size, min_value, max_value):
  len=0
  random_set = set(random.sample(range(min_value, max_value + 1), size))
  subsets = find_subsets(random_set)
    # Print the subsets for the current set
  for subset in subsets:
        len=len+1
  y.append(len)
for i in range(1,11):
  x.append(i)
  size = i
  min_value = 1
  max_value = 10
  create_random_set(size, min_value, max_value)
print(x)
print(y)
plt.plot(x,y)
plt.xlabel('Set size')
plt.ylabel('No. of subsets')
plt.title('Set size vs No. of subsets')
for i, j in zip(x, y):
    plt.text(i, j, f'({i}, {j})', ha='center', va='bottom')

plt.show()
```
